### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = df['mpg'].median()

# Filter cars with mpg above median
filtered_cars = df[df['mpg'] > median_mpg].copy()

# Calculate horsepower-to-weight ratio
filtered_cars['hp_weight_ratio'] = filtered_cars['horsepower'] / filtered_cars['weight']

# Find the maximum ratio value
max_ratio = filtered_cars['hp_weight_ratio'].max()

# Get the cars with the max ratio
best_cars = filtered_cars[filtered_cars['hp_weight_ratio'] == max_ratio]

print(best_cars[['name', 'horsepower', 'weight', 'mpg', 'hp_weight_ratio']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().unstack()
plt.figure(figsize=(10,6))
for origin in avg_mpg.columns:
    plt.plot(avg_mpg.index, avg_mpg[origin], marker='o', label=origin)

plt.title('Average MPG over Years by Origin')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd

# Step 1: Find cars appearing in more than one year
car_year_counts = df.groupby('name')['model_year'].nunique()
multiple_year_cars = car_year_counts[car_year_counts > 1].index

# Step 2 & 3: Filter original data for these cars
filtered_df = df[df['name'].isin(multiple_year_cars)]

# Calculate std deviation, count, and mean mpg for each car
stats = filtered_df.groupby('name')['mpg'].agg(['std', 'count', 'mean']).reset_index()

# Filter for std < 1.0 and count >= 2
consistent_cars = stats[(stats['std'] < 1.0) & (stats['count'] >= 2)]

# Step 4: Find lowest variation(s)
min_std = consistent_cars['std'].min()
lowest_variation_cars = consistent_cars[consistent_cars['std'] == min_std]

# Bonus: Sort by count (descending) then mean mpg (descending)
result = consistent_cars.sort_values(by=['count', 'mean'], ascending=[False, False])

# Display result
print("Consistent cars with low mpg variation:")
print(result[['name', 'count', 'mean']].rename(columns={
    'name': 'Model Name',
    'count': 'Number of Appearances',
    'mean': 'Average MPG'
}))
```
